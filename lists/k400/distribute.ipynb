```python
import matplotlib.pyplot as plt
import numpy as np

list_files = ['trainlist.txt', 'testlist.txt', 'vallist.txt']

labels = {}

# 读取所有文件并统计标签数量
# for name in list_files:
name = 'trainlist.txt'
with open(name, 'r') as f:
    lines = f.readlines()
for line in lines:
    items = line.strip().split(' ')
    if len(items) == 2:
        label_id = int(items[1])
        if label_id not in labels:
            labels[label_id] = 0
        labels[label_id] += 1

print(f"总共找到 {len(labels)} 个不同的标签")
print(f"标签ID范围: {min(labels.keys())} - {max(labels.keys())}")
print(f"总样本数: {sum(labels.values())}")
```

```python
# 生成标签分布直方图
label_ids = sorted(labels.keys())
label_counts = [labels[label_id] for label_id in label_ids]

plt.figure(figsize=(20, 6))
plt.bar(label_ids, label_counts, alpha=0.7, color='skyblue', edgecolor='black', linewidth=0.5)
plt.xlabel('ID')
plt.ylabel('num')
plt.title('hist')
plt.grid(True, alpha=0.3)

# 显示统计信息
plt.text(0.02, 0.98, f'all label: {len(labels)}\nall sample: {sum(labels.values())}\navg label: {sum(labels.values())/len(labels):.1f}', 
         transform=plt.gca().transAxes, verticalalignment='top', 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()

# 显示标签数量的详细统计
print("\n标签数量统计:")
print(f"最少样本数: {min(label_counts)}")
print(f"最多样本数: {max(label_counts)}")
print(f"平均样本数: {np.mean(label_counts):.2f}")
print(f"标准差: {np.std(label_counts):.2f}")
```
